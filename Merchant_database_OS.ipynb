```python
import PySimpleGUI as sg
import os.path
import requests
import cv2
import sqlite3
import os
import shutil
```

```python
sg.theme('Dark')   # 设置当前主题
file_list_column = [
    [
        sg.Text("食品："),
        sg.In(size=(20, 1), enable_events=True, key="category"),
        sg.Button("添加食品类以及价格", key="insert_category")
    ],
    [
        sg.Text("价格："),        
        sg.In(size=(20, 1), enable_events=True, key="value"),
        sg.Button("删除食品类以及价格", key="delete_category")
    ],
    [
        sg.Text("             "),
        sg.Button("显示食品价格清单", key="show_foodvalue"),
        sg.Text(" "),
        sg.Button("更新食品类的价格", key="update_category")
    ]
]
image_viewer_column = [
    [sg.Text("以下为被录入的食品以及价格清单：")],
    [sg.Listbox(values=[], size=(40, 20), key="-POUT-")],
#     [sg.Image(key="-IMAGE-")],
]
layout = [
    [
        sg.Column(file_list_column),
        sg.VSeperator(),
        sg.Column(image_viewer_column),
    ]
]
window = sg.Window("商家食品价格清单系统", layout)
while True:
    event, values = window.read()
    if event == "Exit" or event == sg.WIN_CLOSED:
        break
    if event == "insert_category":
        try:
            food_category = values["category"]
            food_value = values["value"]
            conn = sqlite3.connect("C:\\Users\\Junhao's Alienware\\food_recognition\\Database/name_value.db")
            cursor = conn.cursor()
            cursor.execute("insert into user (id, value) values (?, ?)", (food_category, float(food_value)))
            # 执行查询语句:
            cursor.execute('select * from user')
            # 获得查询结果集:
            values = cursor.fetchall()
            window["-POUT-"].update(values)
            # 关闭Cursor:
            cursor.close()
            # 提交事务:
            conn.commit()
            # 关闭Connection:
            conn.close()
        except:
            pass
    if event == "delete_category":
        try:
            food_category = values["category"]
            food_value = values["value"]
            conn = sqlite3.connect("C:\\Users\\Junhao's Alienware\\food_recognition\\Database/name_value.db")
            cursor = conn.cursor()
            cursor.execute("delete from user where id=?", (food_category,))
            # 执行查询语句:
            cursor.execute('select * from user')
            # 获得查询结果集:
            values = cursor.fetchall()
            window["-POUT-"].update(values)
            # 关闭Cursor:
            cursor.close()
            # 提交事务:
            conn.commit()
            # 关闭Connection:
            conn.close()
        except:
            pass
    if event == "show_foodvalue":
        try:
            food_category = values["category"]
            food_value = values["value"]
            conn = sqlite3.connect("C:\\Users\\Junhao's Alienware\\food_recognition\\Database/name_value.db")
            cursor = conn.cursor()
            # 执行查询语句:
            cursor.execute('select * from user')
            # 获得查询结果集:
            values = cursor.fetchall()
            window["-POUT-"].update(values)
            # 关闭Cursor:
            cursor.close()
            # 提交事务:
            conn.commit()
            # 关闭Connection:
            conn.close()
        except:
            pass
    if event == "update_category":
        try:
            food_category = values["category"]
            food_value = values["value"]
            conn = sqlite3.connect("C:\\Users\\Junhao's Alienware\\food_recognition\\Database/name_value.db")
            cursor = conn.cursor()
            cursor.execute("update user set value=? where id=?", (food_value, food_category))
            # 执行查询语句:
            cursor.execute('select * from user')
            # 获得查询结果集:
            values = cursor.fetchall()
            window["-POUT-"].update(values)
            # 关闭Cursor:
            cursor.close()
            # 提交事务:
            conn.commit()
            # 关闭Connection:
            conn.close()
        except:
            pass

window.close()
```
